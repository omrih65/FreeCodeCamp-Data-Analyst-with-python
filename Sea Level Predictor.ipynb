```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress

def draw_plot():
    # Read data from file
    try:
        data = pd.read_csv("epa-sea-level.csv")
    except:
        print('Erorr')
    # Create scatter plot
    fig, ax = plt.subplots()
    plt.scatter(x = data['Year'], y = data['CSIRO Adjusted Sea Level'])

    # Create first line of best fit
    res = linregress(data['Year'], data['CSIRO Adjusted Sea Level'])
    x_pred = pd.Series(range(1880,2051))
    y_pred = res.intercept + res.slope * x_pred
    plt.plot(x_pred, y_pred ,label='fitted line')
    
    # Create second line of best fit
    data_2000 = data[data[data['Year'] == 2000].index[0] : ]
    res = linregress(data_2000['Year'], data_2000['CSIRO Adjusted Sea Level'])
    x_pred2 = pd.Series(range(2000,2051))
    y_pred2 = res.intercept + res.slope*x_pred2
    plt.plot(x_pred2, y_pred2, label='fitted line')
    # Add labels and title
    ax.set_xlabel('Year')
    ax.set_ylabel('Sea Level (inches)')
    ax.set_title('Rise in Sea Level')
    
    # Save plot and return data for testing (DO NOT MODIFY)
    plt.savefig('sea_level_plot.png')
    return plt.gca()
```
